```python
import pandas as pd
```

Load the raw data

```python
df = pd.read_csv("artist.csv")

print("── Step 1: Load ──────────────────────────────────")
print(f"  Raw shape: {df.shape}  ({df.shape[0]} rows, {df.shape[1]} columns)")
print(f"  Columns: {df.columns.tolist()}")
```

Remove embedded header rows

```python
header_rows = df[df["Artist Name"] == "Artist"]

print("\n── Step 2: Remove embedded header rows ───────────")
print(f"  Found {len(header_rows)} embedded header rows")
print(f"  At raw file positions: {header_rows.index.tolist()}")

df = df[df["Artist Name"] != "Artist"].copy()

print(f"  Shape after removal: {df.shape}")
```

Drop the redundant index column

```python
print("\n── Step 3: Drop redundant index column ───────────")
print(f"  Dropping 'Unnamed: 0' (values were just row numbers 1-1009)")

df = df.drop(columns=["Unnamed: 0"])

print(f"  Columns remaining: {df.columns.tolist()}")
```

Strip comma-formatting and cast numeric columns

```python
print("\n── Step 4: Strip comma-formatting, cast to int64 ─")

for col in ["Lead Streams", "Feats", "Tracks"]:
    before = df[col].dtype
    df[col] = df[col].str.replace(",", "", regex=False).astype("int64")
    print(f"  '{col}': {before} -> {df[col].dtype}")
```

Case 'One Billion' and '100 Million' to int

```python
print("\n── Step 5: Cast 'One Billion' and '100 Million' to int")

for col in ["One Billion", "100 Million"]:
    before = df[col].dtype
    df[col] = df[col].astype(int)
    print(f"  '{col}': {before} -> {df[col].dtype}")
```
