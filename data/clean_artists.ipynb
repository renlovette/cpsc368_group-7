```python
# import libraries
import pandas as pd
```

```python
# load dataset
df = pd.read_csv("raw/artist.csv")

df.head()
```

```python
# check data types
df.dtypes
```

```python
# drop index and last updated column
df = df[["Artist Name", "Lead Streams", "Feats", "Tracks", "One Billion", "100 Million"]]

print(df.columns.tolist())
```

```python
# rename columns to be all lowercase and remove space inbetween words
df.columns = df.columns.str.lower().str.replace(' ', '_')

print(df.columns.tolist())
```

```python
# check for missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# remove commas from lead streams and feats and
df["lead_streams"] = df["lead_streams"].str.replace(',', '')
df["feats"] = df["feats"].str.replace(',', '')
df["tracks"] = df["tracks"].str.replace(',','')
```

```python
print(df.head(3))
```

```python
# found weird values in dataset that need to be removed
print(df[df['lead_streams'] == 'Lead Streams'])

clean_df = df[df['lead_streams'] != 'Lead Streams']
print(clean_df[clean_df['lead_streams'] == 'Lead Streams'])
```

```python
# change data types from string to int
df[["lead_streams", "feats", "tracks", "one_billion", "100_million"]] = clean_df[["lead_streams", "feats", "tracks", "one_billion", "100_million"]].astype(int)

df.head()
```

```python
print(df.describe())
print(df.dtypes)
print(df.shape)
```

```python
# save csv
df.to_csv("processed/artist_clean.csv", index=False)
```
