```python
import pandas as pd

df = pd.read_csv("raw/grammys.csv")
df.head()
```

```python
# select required columns
df = df[["year", "award", "artist", "song_or_album", "winner"]]
df.head()
```

```python
# filter for rows that are from years 2016-2020
df = df[(df["year"] >= 2016) & (df["year"] <= 2020)]
print(df)
```

```python
# rename song_or_album to music and artist to artist_name
df = df.rename(columns={"song_or_album": "music_title",
                       "artist": "artist_name"})

df.head()
```

```python
# standardize music by removing descriptors: delete anything starting with "("
df["music_title"] = df["music_title"].str.split("(").str[0].str.strip()
print(df)
```

```python
# check for missing values
missing_values = df.isnull().sum()
df = df.dropna()

dropped_missing_values = df.isnull().sum()
print(dropped_missing_values)
```
