```python
import pandas as pd

df = pd.read_csv("data/raw/grammys.csv")
df.head()
```

```python
# select required columns
df = df[["year", "award", "artist", "song_or_album", "winner"]]
df.head()
```

```python
# filter for rows that are from years 2016-2024
df = df[(df["year"] >= 2016) & (df["year"] <= 2024)]
print(df)
```

```python
# rename song_or_album to music 
df = df.rename(columns={"song_or_album": "music"})
df.head()
```

```python
# standardize music by removing descriptors: delete anything starting with "("
df["music"] = df["music"].str.split("(").str[0].str.strip()
print(df)
```

```python
# only winners 
df = df[df["winner"] == True]
print(df)
```

```python
# save clean dataset
df.to_csv("data/clean/grammys_clean.csv", index=False)
```
