```python
# load libaries

import pandas as pd
import numpy as np
```

```python
# load datasets

spotify_1 = pd.read_csv("raw/spotify_1.csv")
spotify_2 = pd.read_csv("raw/spotify_2.csv")

# list columns of all three datasets
print("Spotify 1 Columns: \n", spotify_1.columns.tolist(), "\n Spotify 2 Columns:\n", spotify_2.columns.tolist())
```

```python
# renaming columns for standardizing
spotify_1 = spotify_1.rename(columns={"id": "track_id"})
spotify_2 = spotify_2.rename(columns={"id": "track_id"})

spotify_1 = spotify_1.rename(columns={"artist_names": "artist_name"})
spotify_2 = spotify_2.rename(columns={"artists": "artist_name"})

spotify_2 = spotify_2.rename(columns={"name": "track_name"})

print(spotify_1.columns, spotify_2.columns)

print(spotify_1.shape, spotify_2.shape)
```

```python
# CLEANING: Spotify 1 - Spotify Top Songs And Audio Features

# check data types
print(spotify_1.dtypes)

# check for missing values
missing_values_1 = spotify_1.isnull().sum()
print("Missing values:\n", missing_values_1)

# drop missing values
spotify_1 = spotify_1.dropna()

# check for duplicates
spotify_1 = spotify_1.drop_duplicates()
print("Shapes after dropping duplicates:", spotify_1.shape)

spotify_1.sample(5)
```

```python
# summary
spotify_1.describe()
```

```python
# filter for columns related to research question due to database constraints
cleaned_spotify_1 = spotify_1[["track_id", "danceability", "valence", "energy", "tempo", "duration_ms", "streams"]]
print(cleaned_spotify_1.shape)
```

```python
# standardize so all decimals are rounded to 3 decimal points
cleaned_spotify_1[["danceability", "valence", "energy", "tempo"]] = cleaned_spotify_1[["danceability", "valence", "energy", "tempo"]].round(3)

cleaned_spotify_1.head()
```

```python
# CLEANING: Spotify 2 - Spotify Dataset 1921-2020 
# check data types
print(spotify_2.dtypes)

# check for missing values
missing_values_2 = spotify_2.isnull().sum()
print("Missing values:\n", missing_values_2)

# drop missing values
spotify_2 = spotify_2.dropna()

# check for duplicates
spotify_2 = spotify_2.drop_duplicates()
print("Shapes after dropping duplicates:", spotify_2.shape)

spotify_2.sample(5)
```

```python
# summary
spotify_2.describe()
```

```python
# research questions only concerns songs from 2016 onwards, filter for these specific years to meet limit of database
filtered_spotify_2 = spotify_2[spotify_2["year"] >= 2016]

# filter for columns related to research question due to database constraints
cleaned_spotify_2 = filtered_spotify_2[["track_id", "danceability", "valence", "energy", "tempo", "duration_ms", "year", "popularity"]]

# round to 3 decimals
cleaned_spotify_2[["danceability", "valence", "energy", "tempo"]] = cleaned_spotify_2[["danceability", "valence", "energy", "tempo"]].round(3)

cleaned_spotify_2.head()
print(cleaned_spotify_2.shape)
print(cleaned_spotify_2.describe())
```

```python
# check overlap between spotify 1 and spotify 2
overlap = cleaned_spotify_1[cleaned_spotify_1["track_id"].isin(cleaned_spotify_2["track_id"])].shape[0]
print(overlap)
```

```python
spotify_audible_features = cleaned_spotify_1.merge(cleaned_spotify_2[["track_id", "year", "popularity"]], on=['track_id'], how="inner")
print(spotify_audible_features.shape)
print(spotify_audible_features.describe())
```

```python
# save cleaned dataset to csv
spotify_audible_features.to_csv("processed/spotify_audible_features.csv", index=False)
```
