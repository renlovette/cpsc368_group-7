```python
import pandas as pd
import numpy as np
```

```python
# load dataset
df = pd.read_csv("raw/spotify_3.csv")
```

```python
# print columns
print("Spotify 3 Columns: \n", df.columns.tolist())
```

```python
# check data types
print(df.dtypes)
```

```python
# standardize release date to only include years
df = df[df["album_release_date"] != "0000"]
df["release_year"] = pd.to_datetime(df["album_release_date"], format = "mixed").dt.year
```

```python
# filter year between 2016 and 2020
df = df[(df["release_year"] <= 2020) & (df["release_year"] >= 2016)]
```

```python
df.head()
```

```python
# standardize artist_genres

df["artist_genres"] = df["artist_genres"].str.translate({ord('['): None,
                                                        ord(']'): None,
                                                        ord('\''): None})

# check for missing values and empty strings
print(df['artist_genres'].isna().sum(), 'NaN values')
print((df['artist_genres'] == '').sum(), 'empty strings')
```

```python
# remove rows with empty strings
df = df[df["artist_genres"] != '']
```

```python
# filter columns needed
df = df[["track_id", "track_name", "track_popularity", "track_duration_ms", "artist_name", "artist_popularity", "artist_followers", 
                "artist_genres", "album_name", "release_year"]]
```

```python
# save clean dataset
df.to_csv("processed/spotify_track_info.csv", index=False)
```
