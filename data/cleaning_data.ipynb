```python
# load libaries

import pandas as pd
import numpy as np
```

```python
# load datasets

spotify_1 = pd.read_csv("raw/spotify_1.csv")
spotify_2 = pd.read_csv("raw/spotify_2.csv")
spotify_3 = pd.read_csv("raw/spotify_3.csv")

# list columns of all three datasets
print(spotify_1.columns, spotify_2.columns, spotify_3.columns)
```

```python
# renaming columns for merging
spotify_1 = spotify_1.rename(columns={"id": "track_id"})
spotify_3 = spotify_3.rename(columns={"id": "track_id"})

print(spotify_1.columns, spotify_2.columns, spotify_3.columns)

print(spotify_1.shape, spotify_2.shape, spotify_3.shape)
```

```python
# check for overlaps in tracks

shared_between_1_2 = spotify_1[spotify_1["track_id"].isin(spotify_2["track_id"])].shape[0]
print(shared_between_1_2, "tracks overlapping in dataset 1 and 2")

shared_between_1_3 = spotify_1[spotify_1["track_id"].isin(spotify_3["track_id"])].shape[0]
print(shared_between_1_3, "tracks overlapping in dataset 1 and 3")

shared_between_2_3 = spotify_2[spotify_2["track_id"].isin(spotify_3["track_id"])].shape[0]
print(shared_between_2_3, "tracks overlapping in dataset 2 and 3")
```

```python
# merging

spotify_1_2 = spotify_1.merge(spotify_2, on="track_id", how="outer", suffixes=("_1", "_2"))
spotify_1_2_3 = spotify_1.merge(spotify_1_2, on="track_id", how="outer", suffixes=("", "_3"))

print(spotify_1_2_3.shape)
print(spotify_1_2_3.columns)
```
