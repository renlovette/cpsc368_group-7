```python
# load libaries

import pandas as pd
import numpy as np
```

```python
# load datasets

spotify_1 = pd.read_csv("raw/spotify_1.csv")
spotify_2 = pd.read_csv("raw/spotify_2.csv")
spotify_3 = pd.read_csv("raw/spotify_3.csv")

# list columns of all three datasets
print("Spotify 1 Columns: \n", spotify_1.columns.tolist(), "\n Spotify 2 Columns:\n", spotify_2.columns.tolist(), "\n Spotify 3 Columns:\n", spotify_3.columns.tolist())
```

```python
# renaming columns for standardizing
spotify_1 = spotify_1.rename(columns={"id": "track_id"})
spotify_3 = spotify_3.rename(columns={"id": "track_id"})

spotify_1 = spotify_1.rename(columns={"artist_names": "artist_name"})
spotify_3 = spotify_3.rename(columns={"artists": "artist_name"})

spotify_3 = spotify_3.rename(columns={"name": "track_name"})

print(spotify_1.columns, spotify_2.columns, spotify_3.columns)

print(spotify_1.shape, spotify_2.shape, spotify_3.shape)
```

```python
# CLEANING: Spotify 1 - Spotify Top Songs And Audio Features

# check data types
print(spotify_1.dtypes)

# check for missing values
missing_values_1 = spotify_1.isnull().sum()
print("Missing values:\n", missing_values_1)

# check for duplicates
spotify_1 = spotify_1.drop_duplicates()
print("Shapes after dropping duplicates:", spotify_1.shape)

spotify_1.sample(5)
```

```python
# summary
spotify_1.describe()
```

```python
cleaned_spotify_1 = spotify_1[["track_id", "instrumentalness" , "danceability", "liveness", "valence", "loudness", "energy", "acousticness", "tempo", "streams"]]
print(cleaned_spotify_1.shape)
```

```python
# check for overlaps in data
shared_between_1_2 = spotify_1[spotify_1["track_id"].isin(spotify_3["track_id"])].shape[0]
print(shared_between_1_2, "tracks overlapping in dataset 1 and 2")

shared_between_1_3 = spotify_1[spotify_1["track_id"].isin(spotify_2["track_id"])].shape[0]
print(shared_between_1_3, "tracks overlapping in dataset 1 and 3")
```

```python
# CLEANING: Spotify 2 - Spotify Dataset 1921-2020 
```

```python
# check for missing values
missing_values_1 = spotify_1.isnull().sum()
print("Missing values:\n", missing_values_1)

missing_values_2 = spotify_2.isnull().sum()
print("Missing values:\n", missing_values_2)

missing_values_3 = spotify_3.isnull().sum()
print("Missing values:\n", missing_values_3)
```

```python
# Drop rows with missing values
spotify_2 = spotify_2.dropna()
print("Shape after dropping missing values:", spotify_2.shape)
```

```python
# Remove duplicates
spotify_1 = spotify_1.drop_duplicates()
spotify_2 = spotify_2.drop_duplicates()
spotify_3 = spotify_3.drop_duplicates()

print("Shapes after dropping duplicates:", spotify_1.shape, spotify_2.shape, spotify_3.shape)
```

```python
# Check data types
print("\nData types:")
print(spotify_1.dtypes)
print(spotify_2.dtypes)
print(spotify_3.dtypes)
```

```python
# check for overlaps in tracks

shared_between_1_2 = spotify_1[spotify_1["track_id"].isin(spotify_2["track_id"])].shape[0]
print(shared_between_1_2, "tracks overlapping in dataset 1 and 2")

shared_between_1_3 = spotify_1[spotify_1["track_id"].isin(spotify_3["track_id"])].shape[0]
print(shared_between_1_3, "tracks overlapping in dataset 1 and 3")

shared_between_2_3 = spotify_2[spotify_2["track_id"].isin(spotify_3["track_id"])].shape[0]
print(shared_between_2_3, "tracks overlapping in dataset 2 and 3")
```

```python
# merging

spotify_1_2 = spotify_1.merge(spotify_2, on="track_id", how="outer", suffixes=("_1", "_2"))
spotify_1_2_3 = spotify_3.merge(spotify_1_2, on="track_id", how="outer", suffixes=("", "_3"))

print(spotify_1_2_3.shape)
print(spotify_1_2_3.columns)
```

```python
spotify_1_2_3.head()
```

```python
merged_missing_values = spotify_1_2_3.isnull().sum()
print("Missing values:\n", merged_missing_values)

cleaned_spotify = spotify_1_2_3.dropna()
print("Shape after dropping missing values:", cleaned_spotify.shape)
```

```python
cleaned_spotify.describe()
```

```python
cleaned_spotify['is_duplicate'] = cleaned_spotify.duplicated(subset='track_id', keep=False)
print(cleaned_spotify['is_duplicate'].value_counts())
```
