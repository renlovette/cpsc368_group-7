```python
# import libraies 
import pandas as pd
```

```python
# insert artist data 
df = pd.read_csv("artist_clean.csv")

sql = """INSERT INTO artist (
    artist_name, lead_streams, feats, tracks, one_billion, hundred_million) 
    VALUES (
    :1, :2, :3, :4, :5, :6)"""

data = [(
    row["artist_name"],
    row["lead_streams"],
    row["feats"],
    row["tracks"],
    row["one_billion"],
    row["100_million"])
        for _, row in df.iterrows()
       ]

cur = connection.cursor()

cur.executemany(sql, data)
connection.commit()

cur.close()
```

```python
# verify artist data 

cur = connection.cursor()

cur.execute("SELECT COUNT(*) FROM artist")
row_count = cur.fetchone()[0]

print("Number of rows in artist:", row_count)

cur.close()
```

```python
# insert billboard data 
df = pd.read_csv("billboard_clean.csv")

sql = """INSERT INTO billboard (
    track_name, artist_name, peak_rank, weeks_on_board) 
    VALUES (
    :1, :2, :3, :4)"""

data = [(
    row["track_name"],
    row["artist_name"],
    row["peak_rank"],
    row["weeks_on_board"])
        for _, row in df.iterrows()
       ]

cur = connection.cursor()

cur.executemany(sql, data)
connection.commit()

cur.close()
```

```python
# verify billboard data 

cur = connection.cursor()

cur.execute("SELECT COUNT(*) FROM billboard")
row_count = cur.fetchone()[0]

print("Number of rows in artist:", row_count)

cur.close()
```

```python
# insert grammys data (with foreign key adjustments)  

df = pd.read_csv("grammys_clean.csv")

artist_df = pd.read_sql("SELECT artist_name FROM artist", connection)
artist_df.columns = artist_df.columns.str.lower()

valid_artists = set(artist_df["artist_name"])

df_filtered = df[df["artist_name"].isin(valid_artists)]

sql = """INSERT INTO grammys (
    year, award, artist_name, music_title, winner) 
    VALUES (
    :1, :2, :3, :4, :5)"""

data = [(
    row["year"],
    row["award"],
    row["artist_name"],
    row["music_title"],
    row["winner"])
        for _, row in df_filtered.iterrows()
      ]

cur = connection.cursor()

cur.executemany(sql, data)
connection.commit()

cur.close()
```

```python
# verify grammys data 

cur = connection.cursor()

cur.execute("SELECT COUNT(*) FROM grammys")
row_count = cur.fetchone()[0]

print("Number of rows in artist:", row_count)

cur.close()
```
