```python
# ARTIST RELATION CREATION * creations will not rerun; can only be created once  

cur = connection.cursor()
cur.execute("""CREATE TABLE artist (
    artist_name VARCHAR(50) PRIMARY KEY, 
    lead_streams INT, 
    feats INT, 
    tracks INT, 
    one_billion INT, 
    hundred_million INT)"""
           )
connection.commit()
cur.close()
```

```python
# BILLBOARD RELATION CREATION 
cur = connection.cursor()
cur.execute("""CREATE TABLE billboard (
    track_name VARCHAR(50), 
    artist_name VARCHAR(50), 
    peak_rank INT,
    weeks_on_board INT,
    PRIMARY KEY (track_name, artist_name))"""
           )
connection.commit()
cur.close()
```

```python
# BILLBOARD RELATION VERIFICATION 
cur = connection.cursor()

try:
    cur.execute("SELECT * FROM billboard")
    print("Table exists and is accessible.")
except Exception as e:
    print("Table does NOT exist or cannot be accessed:", e)

cur.close()
```

```python
# GRAMMYS RELATION CREATION 
cur = connection.cursor()
cur.execute("""CREATE TABLE grammys (
    year INT, 
    award VARCHAR(50), 
    artist_name VARCHAR(50),
    music_title VARCHAR(50),
    winner VARCHAR(5),
    PRIMARY KEY (year, award, artist_name, music_title),
    FOREIGN KEY (artist_name) REFERENCES artist(artist_name)
    ON DELETE CASCADE)"""
           )
connection.commit()
cur.close()
```

```python
# GRAMMYS RELATION VERIFICATION 
cur = connection.cursor()

try:
    cur.execute("SELECT * FROM grammys")
    print("Table exists and is accessible.")
except Exception as e:
    print("Table does NOT exist or cannot be accessed:", e)

cur.close()
```
