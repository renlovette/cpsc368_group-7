```python
# load libraries 
import pandas as pd
```

```python
# SQL for RQ1: for each peak rank what are the average audio features and average weeks on the chart alng with the number of songs in that rank group

query_1 = """
SELECT
   	b.peak_rank,
   	AVG(b.weeks_on_board) AS avg_weeks_on_board,
    AVG(af.tempo) AS avg_tempo,
  	AVG(af.danceability) AS avg_danceability,
    AVG(af.energy) AS avg_energy,
    AVG(af.valence) AS avg_valence,
    COUNT(*) AS num_songs
FROM track_information ti
JOIN audio_features af
    	ON ti.track_id = af.track_id
JOIN billboard b
    	ON ti.track_name = b.track_name
    	AND ti.artist_name = b.artist_name
GROUP BY b.peak_rank
ORDER BY b.peak_rank
"""

df_1 = pd.read_sql(query_1, connection)
print(df_1)

df_1.to_csv("peak_rank_characteristics_rq1.csv", index = False)
```

```python
# SQL for RQ2: for each award category what is the average amount of spotify_streams along with the number of winners and nominees

query_2 = """
SELECT 
    g.award, g.winner, 
    AVG(af.streams) AS avg_spotify_streams, 
    COUNT(*) AS total_nominations 
FROM 
    grammys g 
JOIN 
    track_information ti 
    ON g.music_title = ti.track_name 
    AND g.artist_name = ti.artist_name 
JOIN 
    audio_features af 
    ON ti.track_id = af.track_id 
GROUP BY 
    g.award, g.winner
"""

df_2 = pd.read_sql(query_2, connection)
print(df_2)

df_2.to_csv("recognition_and_streams_rq2.csv", index = False)
```
