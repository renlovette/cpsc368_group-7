```python
# libraries 

import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# (rq1) peak rank vs average weeks on board

df_1 = pd.read_csv("peak_rank_characteristics_rq1.csv")

plt.figure(figsize=(10, 6))

plt.scatter(df_1["PEAK_RANK"], df_1["AVG_WEEKS_ON_BOARD"], color="royalblue")

plt.title("Peak Rank vs. Average Weeks on Billboard Chart")
plt.xlabel("Peak Rank (1 = highest)")
plt.ylabel("Average Weeks on Chart")

plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# (rq2) average spotify streams vs grammy recognition across award categories 

# Load the data
df_2 = pd.read_csv("recognition_and_streams_rq2.csv")

# Sort by average streams (descending)
df_sorted = df_2.sort_values("AVG_SPOTIFY_STREAMS", ascending=False)

# Plot
plt.figure(figsize=(12, 6))
plt.bar(df_sorted["AWARD"], df_sorted["AVG_SPOTIFY_STREAMS"], color="royalblue")

plt.xticks(rotation=75, ha='right')
plt.ylabel("Average Spotify Streams")
plt.title("Average Spotify Streams by Grammy Award Category")
plt.tight_layout()
plt.show()
```
